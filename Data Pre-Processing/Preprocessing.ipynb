```python
descriptive=pd.DataFrame(index=["mean","median","mode",
                                "Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%",
                               "IQR","1.5rule","Lesser","Greater","min","max"],columns=dataset.columns)
for cName in dataset:
        descriptive[cName]["mean"]=dataset[cName].mean()
        descriptive[cName]["median"]=dataset[cName].median()
        descriptive[cName]["mode"]=dataset[cName].mode()[0]
        
        descriptive[cName]["Q1:25%"]=dataset.describe()[cName]["25%"]
        descriptive[cName]["Q2:50%"]=dataset.describe()[cName]["50%"]
        descriptive[cName]["Q3:75%"]=dataset.describe()[cName]["75%"]
        descriptive[cName]["99%"]=np.percentile(dataset[cName],99)
        descriptive[cName]["Q4:100%"]=dataset.describe()[cName]["max"]
        
        descriptive[cName]["IQR"]=descriptive[cName]["Q3:75%"]-descriptive[cName]["Q1:25%"]
        descriptive[cName]["1.5rule"]=1.5*descriptive[cName]["IQR"]
        descriptive[cName]["Lesser"]=descriptive[cName]["Q1:25%"]- descriptive[cName]["1.5rule"]
        descriptive[cName]["Greater"]= descriptive[cName]["Q3:75%"]+ descriptive[cName]["1.5rule"]
        
        descriptive[cName]["min"]= dataset[cName].min()
        descriptive[cName]["max"]= dataset[cName].max()
```

```python
descriptive
```

```python
lesser =[]
greater =[]

for columnname in dataset.columns:
  # Lesser outlier --> Min value should not be less than lesser
    if (descriptive[columnname]["min"]<descriptive[columnname]["Lesser"]):
        lesser.append(columnname)
  # Greater outlier --> Max value should not be greater than greater.   
    if (descriptive[columnname]["max"]>descriptive[columnname]["Greater"]):
        greater.append(columnname)      
```

```python
lesser,greater
```

```python
for columnname in lesser:
    dataset[columnname][dataset[columnname]<descriptive[columnname]["Lesser"]]=descriptive[columnname]["Lesser"]
for columnname in greater:
    dataset[columnname][dataset[columnname]>descriptive[columnname]["Greater"]]=descriptive[columnname]["Greater"]
```
